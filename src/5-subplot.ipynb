### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Sample data for each subplot
# Line chart data
x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)

# Bar chart data
categories_bar = ['A', 'B', 'C']
values_bar = [15, 30, 45]

# Scatter plot data
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

# Pie chart data
labels_pie = ['Part 1', 'Part 2', 'Part 3', 'Part 4']
sizes_pie = [30, 25, 20, 25]
colors_pie = ['red', 'green', 'blue', 'yellow']

# Create the 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Combined Subplots', fontsize=16)

# Subplot 1: Line chart
axes[0, 0].plot(x_line, y_line)
axes[0, 0].set_title('Line Chart')
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('Y-axis')
axes[0, 0].grid(True)

# Subplot 2: Bar chart
axes[0, 1].bar(categories_bar, values_bar)
axes[0, 1].set_title('Bar Chart')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')

# Subplot 3: Scatter plot
axes[1, 0].scatter(x_scatter, y_scatter)
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X')
axes[1, 0].set_ylabel('Y')

# Subplot 4: Pie chart
axes[1, 1].pie(sizes_pie, labels=labels_pie, colors=colors_pie, autopct='%1.1f%%', startangle=90)
axes[1, 1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axes[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlapping titles
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust rect to make space for suptitle

# Display the plot
plt.show()
```
